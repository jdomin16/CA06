# 2. Splitting data

```python
#Splitting data into x and y
X = data.loc[:, ['IMDB Rating', 'Biography', 'Drama', 'Thriller', 'Comedy', 'Crime', 'Mystery', 'History']]

y = data.loc[:, ['Movie Name']]
```

```python
#Check the data
X.head()
```

```python
y.tail()
```

# 3. Creating a kNN Recommender Model

```python
#Building and fitting the model looking for 5 neighbours
movies = NearestNeighbors(n_neighbors = 5)

movies.fit(X, y)
```

```python
#Input 'The post' movie information
the_post = [7.2, 1, 1, 0, 0, 0, 0, 1]

#Creating a list for similar movies to "The Post"
rec_movies = movies.kneighbors([the_post], return_distance=False)
list_5 = rec_movies.tolist()
movies_recom = list_5[0]
```

```python
#Show the 5 most similar movies to 'the post'
print("Based on your preferences, you may also like to watch:")
for row_num in movies_recom:
  print(data.iloc[row_num]['Movie Name'])
```
